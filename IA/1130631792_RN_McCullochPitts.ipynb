```python
import numpy as np # library to work with arrays
np.random.seed(seed=0) # initializing for random generation
inputs = np.random.choice([0,1], 3)# generate random vector for inputs, sampling from {0,1}
weights = np.random.choice([-1,1], 3) # generate random vector for weights, sampling from {-1,1} 
print(f'Input vector:{inputs}, Weight vector:{weights}')
```

```python
dot = inputs @ weights # dot product operation
print(f'Dot product: {dot}')
```

```python
# Función de activación para AND y OR
def threshold_fireAND_OR(dot: int, thresh: float) -> int:
    if dot >= thresh:
        return 1
    else:
        return 0
```

```python
neuron_thresh = 1
activation = threshold_fireAND_OR(dot, neuron_thresh)
print(f'Activation: {activation}')
```

```python
# Función de activación para XOR
def threshold_fireXOR(dot: int, thresh: float) -> int:
    if dot == thresh:
        return 1
    else:
        return 0
```

# Movie decision table

```python
# build the inputs
input_table = np.array([
    [0,0], # both no
    [0,1], # one no, one yes
    [1,0], # one yes, one no
    [1,1]  # bot yes
])
```

```python
# define some weights
weights = np.array([1,1])
print(f'weights: {weights}')
```

```python
# do the sum
dot_products = input_table @ weights
print(f'Dot products: {dot_products}')
```

El umbral es mayor o igual a 2 ya que acorde a la tabla de verdad y la validación AND se requiere que al menos la suma de los datos ponderados sea igual a 2. Aplica para dos o más entradas. (Se probó con 3 entradas y funcionó), por otra parte los valores usados como pesos son [1,1], por tanto no se esperan cambios en las tablas de verdad.

```python
thresh = 2
for i in range(0,4):
    activation = threshold_fireAND_OR(dot_products[i], thresh)
    print(f'can I watch: {np.where(activation==0,"no movie","a Batman movie!")}')
```

**SOLUCION DEL EJERCICIO "OR"**
**1130631792 - ORLANDO CORREA MARTINEZ** 
El umbral es mayor o igual a 1 ya que acorde a la tabla de verdad y la validación OR se requiere que al menos la suma de los datos ponderados sea igual a 1. Aplica para dos o más entradas. (Se probó con 3 entradas y funcionó), por otra parte los valores usados como pesos son [1,1], por tanto no se esperan cambios en las tablas de verdad. Se comprende que es el caso opuesto al AND

```python
thresh = 1
for i in range(0,4):
    activation = threshold_fireAND_OR(dot_products[i], thresh)
    print(f'can I watch: {np.where(activation==0,"no movie","a Batman movie!")}')
```

**SOLUCION DEL EJERCICIO "XOR"**
**1130631792 - ORLANDO CORREA MARTINEZ** 
El umbral es igual a 1 ya que acorde a la tabla de verdad y la validación XOR se requiere que la suma de los datos ponderados sea igual a 1. Aplica para dos entradas. (Se probó con 3 entradas y no funcionó, esto debido a que ya se genera dos o más umbrales, la prueba con 3 entradas arrojo dos umbrales posibles, 1 y 3), por otra parte los valores usados como pesos son [1,1], por tanto no se esperan cambios en las tablas de verdad. Se comprende que es el caso donde solo si una de las dos condiciones se cumple entonces es válida la respuesta.

```python
thresh = 1
for i in range(0,4):
    activation = threshold_fireXOR(dot_products[i], thresh)
    print(f'can I watch: {np.where(activation==0,"no movie","a Batman movie!")}')
```

**SOLUCION DEL EJERCICIO "NOT"**
**1130631792 - ORLANDO CORREA MARTINEZ** Para este caso tenemos en cuenta el significado de la compuerta Logica NOT, que es el inverso de un valor entregado, para ser eficientes con el trabajo ya realizado, se propone alterar otro parámetro de la función de activación, es decir, el evaluador final de cada función, actualmente todo valor es evaluado en 0, de ser así no hay película, si se cambia a 1 toda la respuesta final se invierte. Vamos a calcular a continuación cada caso de compuerta Logica descrita en este ejercicio y se imprimirán los resultados.

```python
#Compuerta NOT_AND
thresh = 2
for i in range(0,4):
    activation = threshold_fireAND_OR(dot_products[i], thresh)
    print(f'can I watch: {np.where(activation==1,"no movie","a Batman movie!")}')
```

```python
#Compuerta NOT_OR
thresh = 1
for i in range(0,4):
    activation = threshold_fireAND_OR(dot_products[i], thresh)
    print(f'can I watch: {np.where(activation==1,"no movie","a Batman movie!")}')
```

```python
#Compuerta NOT_XOR
thresh = 1
for i in range(0,4):
    activation = threshold_fireXOR(dot_products[i], thresh)
    print(f'can I watch: {np.where(activation==1,"no movie","a Batman movie!")}')
```
