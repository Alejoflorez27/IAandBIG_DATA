#Python code

```python
import numpy as np # library to work with arrays
np.random.seed(seed=0) # initializing for random generation
inputs = np.random.choice([0,1], 3)# generate random vector for inputs, sampling from {0,1}
weights = np.random.choice([-1,1], 3) # generate random vector for weights, sampling from {-1,1} 
print(f'Input vector:{inputs}, Weight vector:{weights}')
```

```python
dot = inputs @ weights # dot product operation
print(f'Dot product: {dot}')
```

```python
# we need to activate only when the sum surpass the threshold
def threshold_fire(dot: int, thresh: float) -> int:
    if dot >= thresh:
        return 1
    else:
        return 0
```

I know it is a silly function that can be implemented in a single line, but later it would become more complex

```python
neuron_thresh = 1
activation = threshold_fire(dot, neuron_thresh)
print(f'Activation: {activation}')
```

```python
hard_thresh = 3
activation = threshold_fire(dot, hard_thresh)
print(f'Activation: {activation}')
```

# Movie decision table

---
| PopCorn | Time | Watch Movie |
| :-- | :-- | :-- |
| 0 | 0 | 0 |
| 1 | 0 | 0 |
| 0 | 1 | 0 |
| 1 | 1 | 1 |

Only when I have popcorn and time I can watch a movie

```python
# build the inputs
input_table = np.array([
    [0,0], # both no
    [0,1], # one no, one yes
    [1,0], # one yes, one no
    [1,1]  # bot yes
])
```

```python
# define some weights
weights = np.array([1,1])
print(f'weights: {weights}')
```

```python
# do the sum
dot_products = input_table @ weights
print(f'Dot products: {dot_products}')
```

RESULTADOS POR TABLA DE VERDAD AND

```python
#compuerta logica AND por medio del umbral de 2 y el producto punto es el mismo para el resto de compuertas
thresh = 2
for i in range(0,4):
    activation = threshold_fire(dot_products[i], thresh)
    print(f'can I watch: {np.where(activation==0,"no movie","a Disney movie!")}')
```

Tabla de la verdad NOT AND
---
| PopCorn | Time | Watch Movie |
| :-- | :-- | :-- |
| 0 | 0 | 1 |

```python
# we need to activate only when the sum surpass the threshold
def threshold_fire_Not_OR_AND(dot: int, thresh: float) -> int:
    if dot >= thresh:
        return 0
    else:
        return 1
```

```python
#compuerta logica NOT AND por medio del umbral de 2 y el producto punto es el mismo para el resto de compuertas
thresh = 2
for i in range(0,4):
    activation = threshold_fire_Not_OR_AND(dot_products[i], thresh)
    print(f'can I watch: {np.where(activation==0,"no movie","a Disney movie!")}')
```

FUNCION PARA OBTENER XOR CUANDO EL PRODUCTO PUNTO ES IGUAL AL UMBRAL

```python
# we need to activate only when the sum surpass the threshold
def threshold_fire(dot: int, thresh: float) -> int:
    if dot == thresh:
        return 1
    else:
        return 0
```

```python
# we need to activate only when the sum surpass the threshold
def threshold_fireNotXor(dot: int, thresh: float) -> int:
    if dot != thresh:
        return 1
    else:
        return 0
```

```python
#compuerta logica NOT XOR por medio del umbral de 1 y el producto punto es el mismo para el resto de compuertas
thresh = 1
for i in range(0,4):
    activation = threshold_fireNotXor(dot_products[i], thresh)
    print(f'can I watch: {np.where(activation==0,"no movie","a Disney movie!")}')
```
