```python
class GenericTranslator:

  def __init__(self):
    pass

  @classmethod
  def metodo1(self):
    pass

  @classmethod
  def metodo2(self):
    pass

class EnglishTranslator(GenericTranslator):

  def __init__(self):
      self.translations = {"papá":"dad", "mamá":"mom", "hermana":"sister"}
  
  def translate(self,msg):
      return self.translations.get(msg)

class FrenchTranslator(GenericTranslator):

  def __init__(self):
      self.translations = {"papá":"pére", "mamá":"mére", "hermana":"Schwerter"}
  
  def translate(self,msg):
      return self.translations.get(msg)

class GermanTranslator(GenericTranslator):

  def __init__(self):
    self.translations = {"papá":"papa", "mamá":"mama", "hermano":"Schwerter"}
  
  def translate(self,msg):
      return self.translations.get(msg)

def Factory(language = "English"):
    translators = {
        "French":FrenchTranslator,
        "English":EnglishTranslator,
        "German":GermanTranslator
    }
    return translators[language]()

if __name__ == "__main__":  
  traductorIngles = Factory("English")
  traductorFrances = Factory("French")
  traductorGerman = Factory("German")

  print(f"papá en Ingles: {traductorIngles.translate('papá')}")
  print(f"papá en Aleman: {traductorGerman.translate('papá')}")
  print(f"papá en Frances: {traductorFrances.translate('papá')}")
```
