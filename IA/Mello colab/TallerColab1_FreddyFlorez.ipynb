**Importar las librerias**

```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

 **Preparacion del Dataset**

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()
#Escalamiento de dataset
x_train, x_test = x_train /255.0, x_test / 255.0
```

**Tama√±o del dataset tanto de entrenamiento como para pruebas**

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape,
```

Manejo de graficas con matplolit

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

plt.figure(figsize=(10,10))


#plt.title("Freddy Florez")
for i in range(25):
    #plt.title("Freddy Florez")
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_train[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[y_train[i]])
#plt.title("Freddy Florez", position=(-1,3.5),
#          fontdict={'family': 'serif', 
#                   'color' : 'darkblue',
#                   'weight': 'bold',
#                   'size': 18})
#plt.title("Freddy Florez")
plt.show()
```

**Creacion del modelo de la red neuronal**

```python
model = tf.keras.models.Sequential(
  [
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    #Nodos de procesamiento
    tf.keras.layers.Dense(128, activation="relu"),
    #(10 nodos-10prendas)
    tf.keras.layers.Dense(10),
  ]
)
```

**Presente arquitectura del modelado**
ver imagen de arquitectura

```python
tf.keras.utils.plot_model(
  model,
  "/tmp/model.png",
  show_shapes=True,
)
```

```python
predictions = model(x_train[:1]).numpy()
predictions
```

```python
# softmax = tf.exp(logits) / tf.reduce.sum(tf.exp(logits, axis)
tf.nn.softmax(predictions).numpy()

```
