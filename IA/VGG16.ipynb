//Iniciamos cargando todas las librerias necesarias para la ejecucion del proceso, usamos la libreria tensorflow, junto con la aplicacion de la libreria keras para el preprocesamiento de las imagenes y las instrucciones de vgg16 para entrenar nuestra red y que su procedimiento se ejecute de manera rapida

```python
# %tensorflow_version 2.x
import tensorflow as tf
from tensorflow import keras
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.applications.vgg16 import preprocess_input
from keras.applications.vgg16 import decode_predictions
from keras.applications.vgg16 import VGG16
import numpy as np
import matplotlib.pyplot as plt
print(tf.__version__)
print(keras.__version__)
```

Set the model to VGG16 and print model summary.

En la siguiente instruccion observamos que carga el modelo de la version, observamos que carga en un proceso de 3 segundo todas las caracteristicas e instrucciones necesarias para procesar nuestra red. La velocidad con la que cargo nos permite evidenciar que ejecutara los procesos de manera instantanea

```python
model = VGG16()
```

```python
print(model.summary())
```

```python
image=tf.keras.utils.get_file(origin='https://www.google.com/url?sa=i&url=https%3A%2F%2Fwww.muyinteresante.es%2Fmascotas%2Ffotos%2Fimpresionantes-curiosidades-sobre-perros-que-fuerte&psig=AOvVaw0-FYMX2SvYcspTFlE6_fWW&ust=1653263790894000&source=images&cd=vfe&ved=0CAwQjRxqFwoTCJDY59nl8fcCFQAAAAAdAAAAABAD')
```

```python
image = load_img(image, target_size=(224, 224))
```

```python
plt.imshow(image)
```

Convert image to Numpy array

```python
image = tf.keras.preprocessing.image.img_to_array(image)
```

Reshape data for the model (change the shape of the input).

```python
image.shape[0], image.shape[1], image.shape[2]
```

```python
image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
```

Preprocess data for the VGG model.

```python
image = preprocess_input(image)
```

```python
image
```

Predict the probability across all output classes.

```python
p = model.predict(image)
```

Convert the probabilities to class labels.

```python
label = decode_predictions(p)
```

```python
print(label)
```
